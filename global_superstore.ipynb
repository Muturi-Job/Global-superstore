```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Global Superstore.txt', sep='\t')
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.drop(columns=['记录数', 'Row ID', 'Market2'], errors='ignore')
```

```python
df.shape
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df['delivery_time_days'] = (df['ship_date'] - df['order_date']).dt.days
df['order_month'] = df['order_date'].dt.month
df['order_year'] = df['order_date'].dt.year
df['order_day'] = df['order_date'].dt.day_name()
df['profit_margin'] = np.where(df['sales'] != 0, df['profit'] / df['sales'], 0)
```

```python
df.columns
```

```python
df.to_csv('Global_Superstore_Cleaned.csv', index=False)
print("Data Cleaned and Exported successfully!")
```
